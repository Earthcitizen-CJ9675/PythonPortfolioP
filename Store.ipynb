# Monthly Trend on Sales and Profit (Line Chart)

```python
import pandas as pd 
import matplotlib.pyplot as plt 

# Read the CSV file with proper encoding and date parsing
df = pd.read_csv('C:/Users/JQE9675M2SSD512/[DeskPy/GlobalSuperstore/Store.csv', 
                 encoding='ISO-8859-1', 
                 parse_dates=['Order Date'], 
                 dayfirst=True)

# Standardize the 'Order Date' column
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)

# Convert Sales and Profit columns to numeric (before resampling), in case they contain non-numeric values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Set 'Order Date' as the index
df.set_index('Order Date', inplace=True)

# Resample the data to monthly totals using 'M' (month-start or month-end as needed)
monthly_totals = df.resample('ME').agg({
    'Sales': 'sum',
    'Profit': 'sum'
})

# Format the index to 'Month-Year'
monthly_totals.index = monthly_totals.index.to_period('M').strftime('%b-%Y')

# Plot the monthly totals as a line chart
plt.figure(figsize=(20, 6))

plt.plot(monthly_totals.index, monthly_totals['Sales'], label='Monthly Sales', marker='o', color='blue')
plt.plot(monthly_totals.index, monthly_totals['Profit'], label='Monthly Profit', marker='o', color='green')

# Add labels and title
plt.title('Monthly Sales and Profit') 
plt.xlabel('Month-Year') 
plt.ylabel('Amount') 
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
#plt.grid(True)
plt.grid(axis='x')  # Grid only for the x-axis (vertical lines)

# Add legend and adjust layout
plt.legend()
plt.tight_layout()  # Adjust layout for better display
plt.show()
```

# Sales and Profits Trend of Technology, Furniture, and Office Supplies

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with proper encoding and date parsing
df = pd.read_csv('C:/Users/JQE9675M2SSD512/[DeskPy/GlobalSuperstore/Store.csv', 
                 encoding='ISO-8859-1', 
                 parse_dates=['Order Date'], 
                 dayfirst=True)

# Standardize the 'Order Date' column
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)

# Convert Sales and Profit columns to numeric (before resampling), in case they contain non-numeric values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Extract month and year from 'Order Date' for grouping
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# Group the data by YearMonth and Category, summing up Sales and Profit
monthly_trends = df.groupby(['YearMonth', 'Category'])[['Sales', 'Profit']].sum().reset_index()

# Plot the trends for each category
categories = ['Technology', 'Furniture', 'Office Supplies']

for category in categories:
    subset = monthly_trends[monthly_trends['Category'] == category]
    
    plt.figure(figsize=(20, 6))
    
    # Plot Sales
    plt.plot(subset['YearMonth'].astype(str), subset['Sales'], label='Sales', marker='o', color='blue')
    
    # Plot Profit
    plt.plot(subset['YearMonth'].astype(str), subset['Profit'], label='Profit', marker='o', color='green')
    
    plt.title(f'Monthly Trend of Sales and Profit for {category}')
    plt.xlabel('Month-Year')
    plt.ylabel('Amount')
    plt.xticks(rotation=45)
    plt.grid(axis='x')  # Grid only for the x-axis (vertical lines)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# Sales Trend of Subcategories under Technology

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with proper encoding and date parsing
df = pd.read_csv('C:/Users/JQE9675M2SSD512/[DeskPy/GlobalSuperstore/Store.csv', 
                 encoding='ISO-8859-1', 
                 parse_dates=['Order Date'], 
                 dayfirst=True)

# Convert 'Order Date' to datetime. This allows us to resample the data by month.
# Standardize the 'Order Date' column
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)

# Convert Sales and Profit columns to numeric (before resampling), in case they contain non-numeric values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Set 'Order Date' as the index 
#df.set_index('Order Date', inplace=True)---incorporate this with the filter

# Filter data to include only the 'Technology' category
tech_data = df[df['Category'] == 'Technology']

# Group by 'Sub-Category' and resample by month to sum the 'Sales' values
tech_data.set_index('Order Date', inplace=True)
monthly_sales = tech_data.groupby('Sub-Category').resample('ME')['Sales'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(30, 6))

# Plot each sub-category's sales over time
for sub_category in monthly_sales['Sub-Category'].unique():
    sub_category_data = monthly_sales[monthly_sales['Sub-Category'] == sub_category]
    plt.plot(sub_category_data['Order Date'], sub_category_data['Sales'], label=sub_category)

# Add labels, title, and legend
plt.title('Monthly Sales of Technology Sub-Categories')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Sub-Category')

# Show the plot
plt.tight_layout()
plt.show()
```

# Sales Trend for Phones (Line Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with proper encoding and date parsing
df = pd.read_csv('C:/Users/JQE9675M2SSD512/[DeskPy/GlobalSuperstore/Store.csv', 
                 encoding='ISO-8859-1', 
                 parse_dates=['Order Date'], 
                 dayfirst=True)

# Convert 'Order Date' to datetime. This allows us to resample the data by month.
# Standardize the 'Order Date' column
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)

# Convert Sales and Profit columns to numeric (before resampling), in case they contain non-numeric values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Filter data to include only the 'Phones' sub-category under 'Technology'
phones_data = df[(df['Category'] == 'Technology') & (df['Sub-Category'] == 'Phones')]

# Set 'Order Date' as the index and resample by month, summing 'Sales' and 'Profit'
phones_data.set_index('Order Date', inplace=True)
monthly_sales_profit = phones_data.resample('ME')[['Sales', 'Profit']].sum().reset_index()

# Create the line plot
plt.figure(figsize=(20, 6))

# Plot monthly sales trend
plt.plot(monthly_sales_profit['Order Date'], monthly_sales_profit['Sales'], label='Sales', color='blue', marker='o')

# Plot monthly profit trend
plt.plot(monthly_sales_profit['Order Date'], monthly_sales_profit['Profit'], label='Profit', color='green', marker='o')

# Add labels, title, and legend
plt.title('Monthly Sales and Profit Trends for Phones')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

# Total Categorical Sales (Tree Map)

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify  # for treemap

# Load your data from the CSV file
df = pd.read_csv('C:/Users/JQE9675M2SSD512/[DeskPy/GlobalSuperstore/Store.csv', encoding='ISO-8859-1')

# Convert Sales and Profit columns to numeric (before resampling), in case they contain non-numeric values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Group by 'Category' and sum the 'Sales' values
grouped_data = df.groupby(['Category'])['Sales'].sum().reset_index()

# Create the treemap
plt.figure(figsize=(12, 8))
squarify.plot(sizes=grouped_data['Sales'], label=grouped_data['Category'], alpha=0.8)

# Add labels and title
plt.title('Total Sales by Category')
plt.axis('off')  # Hide axes

# Show the plot
plt.show()
```

# Sales of Sub-categories in Different Markets

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data from the CSV file
df = pd.read_csv('C:/Users/JQE9675M2SSD512/[DeskPy/GlobalSuperstore/Store.csv', 
                 encoding='ISO-8859-1', 
                 parse_dates=['Order Date'], 
                 dayfirst=True)

# Convert Sales and Profit columns to numeric (before resampling), in case they contain non-numeric values
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# List of markets
markets = df['Market'].unique()

# Plot a separate chart for each market
for market in markets:
    # Filter data for the current market
    market_data = df[df['Market'] == market]
    
    # Group by Sub-Category and sum Sales
    subcategory_sales = market_data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
    
    # Create the plot
    plt.figure(figsize=(10, 3))
    
    # Bar chart
    plt.bar(subcategory_sales.index, subcategory_sales.values, color='skyblue')
    
    # Add labels and title
    plt.title(f'Total Sales of Sub-Categories in {market}', fontsize=14)
    plt.xlabel('Sub-Category', fontsize=12)
    plt.ylabel('Total Sales', fontsize=12)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Show the plot
    plt.tight_layout()
    plt.show()
```
